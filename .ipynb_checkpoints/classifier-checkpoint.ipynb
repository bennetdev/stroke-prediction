# Data Cleaning

```python
data.info()
```

```python
data.isnull().sum()
```

Drop all rows not containing a bmi value seems to be the best and easiest method in this case

```python
data = data.dropna(axis=0)
```

I also drop the instance containing "Other" as gender. This is just because its only one person out of 4909 and it therefore would not be a predictive factor anyway.

```python
data = data.drop(data[data.gender=="Other"].index, axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop("stroke", axis=1), data.stroke, test_size=0.20, random_state=42)
```

```python
ord_attributes = ["hypertension", "heart_disease", "ever_married", "Residence_type"]
cat_attributes = ["work_type", "smoking_status", "gender"]
num_attributes = ["age", "avg_glucose_level", "bmi"]

col_transformer = ColumnTransformer([
    ("ord", OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan), ord_attributes),
    ("cat", OneHotEncoder(), cat_attributes),
    ("num", StandardScaler(), num_attributes)
])
```

```python
models = [DecisionTreeClassifier(), KNeighborsClassifier(), LogisticRegression()]
columns = ["Klassifikator", "Cross-Evaluation-Mean", "Test-Accuracy"]
results = pd.DataFrame(columns=columns)

for model in models:
    name = model.__class__.__name__
    pipeline = make_pipeline(col_transformer, model)
    train_score = cross_val_score(pipeline, X_train, y_train, cv=10).mean()
    
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    test_score = accuracy_score(y_test, y_pred)
    results = pd.concat([results, pd.DataFrame([[name ,train_score, test_score]], columns=columns)], ignore_index=True, axis=0)
results
```

LogisticRegression is the best performing model in this case. This is true for both the Cross Evaluation and Test-Accuracy.

```python
pipeline = make_pipeline(col_transformer, LogisticRegression())
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
accuracy_score(y_test, y_pred)
```

# optimize hyperparameter

```python
from sklearn.model_selection import GridSearchCV
pipeline.get_params()
```

```python
grid = {'logisticregression__C': [0.01,0.1,0.5,1,1.5,5,10,100],
        'logisticregression__max_iter': [400],
        'logisticregression__penalty': ["l1", "l2", "elasticnet"]}
gs = GridSearchCV(pipeline, grid, cv=10, verbose=0)
gs.fit(X_train, y_train)
```

```python
print(f"Scores: {gs.cv_results_['mean_test_score']}")
print(f"Bester Parameter: {gs.best_params_}")
print(f"Bester Score: {gs.best_score_}")
y_pred = gs.best_estimator_.predict(X_test)
```

The model did not improve in relation to the initial model.  
I managed to create a simple LogisticRegression Model with an accuracy of 96.05%
